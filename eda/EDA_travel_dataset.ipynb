```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

file_path = Path('../data_acquisition/raw/traveldata-export.xlsx')
df = pd.read_excel(file_path)
df.head()
```

## Basic Overview

```python
df.info()
df.describe(include='all')
```

## Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.head(20)
```

## Data Cleaning

```python
# Example cleaning steps
df['date'] = pd.to_datetime(df['date'], errors='coerce')

print(df.shape)

# Drop duplicates
df = df.drop_duplicates()

print(df.shape)
```

## Key Distributions

```python
df['cost_CHF'].hist()
plt.title('Distribution of Travel Costs')
plt.xlabel('CHF')
plt.ylabel('Frequency')
plt.show()
```

## Grouped Analysis

```python
df.columns.tolist()
```

```python
df.groupby('transport_mode')['cost_CHF'].mean().sort_values().plot(kind='bar')
plt.title('Average Cost by Transport Type')
plt.show()
```
